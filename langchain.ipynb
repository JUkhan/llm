```python
from langchain.agents import create_sql_agent
from langchain.agents.agent_toolkits import SQLDatabaseToolkit
from langchain.sql_database import SQLDatabase
from langchain.agents import AgentExecutor
from langchain.agents.agent_types import AgentType
#from langchain.chat_models import ChatOpenAI
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_community.llms import Ollama
llm=Ollama(model="llama3")
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import chat_models
from langchain.output_parsers import PydanticOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_core.pydantic_v1 import BaseModel, Field, validator
import os
print(os.environ.get('GROQ_API_KEY'))
chat = ChatGroq(temperature=0, groq_api_key=os.environ.get('GROQ_API_KEY'), model_name="mixtral-8x7b-32768")
system = "You are a helpful assistant."
human = "{text}"
prompt = ChatPromptTemplate.from_messages([("system", system), ("human", human)])

chain = prompt | chat
chain.invoke({"text": "Explain the importance of low latency LLMs."})
```

```python
#create sql db and toolkit
db = SQLDatabase.from_uri("sqlite:///Chinook.db")
toolkit = SQLDatabaseToolkit(db=db, llm=llm)
```

```python
#create agent
agent= create_sql_agent(
    llm=llm,
    toolkit=toolkit,
    verbose=True,
    agent_type=AgentType.OPENAI_FUNCTIONS
)
```

```python
#describing a table
print(agent.invoke("Describe the playlisttrack table"))
```

```python
#running queries
print(agent.invoke(
    "List the total sales per country. Which country's customers spent the most?"
))
```
